# 🏔️ Terraform Quiz

Bienvenue dans l'atelier de pratique Terraform !

**Comment ça marche :**
- Pour les **QCM** : modifie la variable `reponse` avec la lettre de ton choix (`"A"`, `"B"`, `"C"` ou `"D"`), puis exécute la cellule avec `Shift+Entrée`
- Pour les **exercices de code** : remplis la variable `ma_reponse` avec ta réponse, puis exécute
- Le résultat s'affiche immédiatement ✅ ou ❌

---

````python
# Cellule utilitaire — exécute cette cellule en premier (Shift+Entrée)
score = {"correct": 0, "total": 0}

def check(reponse, correct, explication):
    score["total"] += 1
    ok = reponse.strip().upper() == correct.strip().upper()
    if ok:
        score["correct"] += 1
        print(f"✅ Correct ! ({score['correct']}/{score['total']})")
    else:
        print(f"❌ Pas tout à fait. La bonne réponse était : {correct}")
    print(f"💡 {explication}")

def score_final():
    pct = round(score['correct'] / score['total'] * 100) if score['total'] else 0
    emoji = "🎉" if pct == 100 else "🚀" if pct >= 80 else "👍" if pct >= 60 else "📖"
    print(f"{emoji} Score final : {score['correct']}/{score['total']} ({pct}%)")

print("✅ Prêt ! Lance les exercices.")
````
